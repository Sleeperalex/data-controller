```python
import spacy
import pandas as pd

# spaCy's English NLP model
nlp = spacy.load("en_core_web_sm")

df3 = pd.read_csv('controverses.csv', delimiter=';')
df3 = df3[:100]

# Preprocess 'SUMMARY_V1' column
df3['SUMMARY_V1'] = df3['SUMMARY_V1'].fillna("")
df3['SUMMARY_V1'] = df3['SUMMARY_V1'].astype(str)

def extract_keywords(text):
    doc = nlp(text)
    return [token.text for token in doc if token.is_alpha and not token.is_stop]

df3['KEYWORDS'] = df3['SUMMARY_V1'].apply(extract_keywords)

df3[['SUMMARY_V1', 'KEYWORDS']][:30]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Combine all keywords into a single string per document
df3['KEYWORDS_STR'] = df3['KEYWORDS'].apply(lambda x: ' '.join(x))

# Compute TF-IDF weights
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df3['KEYWORDS_STR'])

# Convert TF-IDF matrix to DataFrame for readability
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Add a column for weighted keywords (as a dictionary per document)
df3['WEIGHTED_KEYWORDS'] = tfidf_df.apply(lambda row: {vectorizer.get_feature_names_out()[i]: weight
                                                       for i, weight in enumerate(row) if weight > 0}, axis=1)
pd.set_option("display.max_colwidth", None)

df3[['SUMMARY_V1', 'KEYWORDS', 'WEIGHTED_KEYWORDS']][:30]
```

```python
# filtered by weight > 0.2
df3['WEIGHTED_KEYWORDS'] = tfidf_df.apply(lambda row: {vectorizer.get_feature_names_out()[i]: weight
                                                       for i, weight in enumerate(row) if weight > 0.2}, axis=1)

df3[['SUMMARY_V1', 'KEYWORDS', 'WEIGHTED_KEYWORDS']][:30]
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
import pandas as pd

nltk.download('vader_lexicon')

df4 = pd.read_csv('controverses.csv', delimiter=';')
df4 = df4[:100]

# Preprocess the 'SUMMARY_V1' column
df4['SUMMARY_V1'] = df4['SUMMARY_V1'].fillna("")
df4['SUMMARY_V1'] = df4['SUMMARY_V1'].astype(str)


# VADER Sentiment Analysis
def extract_sentiment_vader(text):
    sid = SentimentIntensityAnalyzer()
    sentiment_scores = sid.polarity_scores(text)
    return (f"Phrase: {text}: Polarity: {sentiment_scores['compound']}, Subjectivity: {sentiment_scores['neu']}")


df4['SENTIMENT_VADER'] = df4['SUMMARY_V1'].apply(extract_sentiment_vader)
```

```python
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification
import torch
import numpy as np

# Load tokenizer and model
tokenizer = DistilBertTokenizer.from_pretrained('descartes100/distilBERT_ESG')
model = DistilBertForSequenceClassification.from_pretrained('descartes100/distilBERT_ESG')

# Input text
#text = "comp_V is involved in controversies related to its Labor Rights, Human Rights and Customers impacts. It faces significant concerns related to Supply Chain Labor Standards."
#text = "comp_SAKY does not face any ESG controversies."
#text= "comp_TG is involved in controversies related to its Labor Rights and Customers impacts. It faces significant concerns related to Collective Bargaining & Union."
text= "comp_Y is involved in controversies related to its Customers, Governance, Environment and Labor Rights impacts.It faces significant concerns related to Product Safety & Quality, Bribery & Fraud, and Customer Relations."

#text="Political corruption continues to plague the government, with several high-ranking officials involved in bribery scandals. The lack of transparency in decision-making processes has led to widespread public distrust. In addition, human rights abuses have been reported, particularly against minority groups, further tarnishing the nation's reputation on the global stage. The government's failure to address these issues has exacerbated social inequality, leading to protests and a significant erosion of public confidence."

#text="The company has been repeatedly criticized for its lack of accountability and transparency in governance practices. Numerous reports highlight the mismanagement of resources, with executives prioritizing profits over ethical considerations. Socially, the company has faced severe backlash for its exploitation of labor in developing countries, with employees working under unsafe conditions and receiving inadequate compensation. Environmentally, the company’s operations have been linked to significant deforestation, contributing to the destruction of critical ecosystems and worsening climate change. These ongoing issues have led to mounting public outrage and calls for greater corporate responsibility."



# Set up device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

# Tokenize and prepare input
encoding = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
encoding = {k: v.to(device) for k, v in encoding.items()}

# Perform inference
outputs = model(**encoding)
logits = outputs.logits

# Apply sigmoid to get probabilities
sigmoid = torch.nn.Sigmoid()
probs = sigmoid(logits.squeeze().cpu())

# Create predictions (binary classification for each category)
predictions = np.zeros(probs.shape)
predictions[np.where(probs >= 0.4)] = 1

# Define label mappings
categories = ["Environmental", "Social", "Governance"]
labels = ["Negative", "Neutral", "Positive"]

# Print results
for i, category in enumerate(categories):
    print(f"{category}:")
    for j, label in enumerate(labels):
        print(f"  {label}: {probs[i * len(labels) + j]:.2f} ({'Selected' if predictions[i * len(labels) + j] == 1 else 'Not Selected'})")
```
