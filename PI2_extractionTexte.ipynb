```python
block_dict = {}

page_num = 1

for page in doc: # Iterate all pages in the document

      file_dict = page.get_text('dict') # Get the page dictionary

      block = file_dict['blocks'] # Get the block information

      block_dict[page_num] = block # Store in block dictionary

      page_num += 1 # Increase the page value by 1
```

```python
import re

spans = pd.DataFrame(columns=['xmin', 'ymin', 'xmax', 'ymax', 'text', 'tag'])

rows = []

for page_num, blocks in block_dict.items():
    for block in blocks:

        if block['type'] == 0:
            for line in block['lines']:
                for span in line['spans']:
                    xmin, ymin, xmax, ymax = list(span['bbox'])

                    font_size = span['size']

                    text = unidecode(span['text'])

                    span_font = span['font']

                    is_upper = False

                    is_bold = False

                    if "bold" in span_font.lower():

                        is_bold = True

                    if re.sub("[\(\[].*?[\)\]]", "", text).isupper():

                        is_upper = True

                    if text.replace(" ","") !=  "":

                        rows.append((xmin, ymin, xmax, ymax, text, is_upper, is_bold, span_font, font_size))


span_df = pd.DataFrame(rows, columns=['xmin','ymin','xmax','ymax', 'text', 'is_upper','is_bold','span_font', 'font_size'])
```

We will define three types of tag: h, p, and s.

    The ‘h’ tag denotes the text which is bigger and more important than normal paragraphs. The text with the ‘h’ tag is usually in the UPPER CASE and has the bold style.
    The ‘p’ tag stands for paragraph, or the normal content in the document. We can find the proper text with ‘p’ tag by counting the number of occurrences of each text size in a document, and then choose the text size which occurs most.
    The ‘s’ tag will be used for less important text, which is smaller than ‘p’ text.


```python
import numpy as np

span_scores = []
span_num_occur = {}
special = '[(_:/,#%\=@)]'

for index, span_row in span_df.iterrows():
    score = round(span_row.font_size)
    text = span_row.text

    if not re.search(special, text):
        if span_row.is_bold:
            score +=1

        if span_row.is_upper:
            score +=1

    span_scores.append(score)

values, counts = np.unique(span_scores, return_counts=True)
```

# Count number of unique text styles

```python
values, counts = np.unique(span_scores, return_counts=True)

style_dict = {}

for value, count in zip(values, counts):

    style_dict[value] = count

sorted(style_dict.items(), key=lambda x: x[1])
```

```python
#paragraph size most common
p_size = max(style_dict, key=style_dict.get)

idx = 0
tag = {}

for size in sorted(style_dict.keys(), reverse=True):  # Sort descending
    idx += 1
    if size == p_size:
        idx = 0
        tag[size] = 'p'  # Assign paragraph tag
    elif size > p_size:
        tag[size] = f'h{idx}'  # Assign heading tag
    else:
        tag[size] = f's{idx}'  # Assign small text tag
```

```python
import pandas as pd
import Levenshtein

def remove_similar_text_fuzzy(df, text_column, threshold=90):
    unique_texts = []
    indexes_to_remove = []
    similarity_data = []

    for index, text in df[text_column].items():
        for unique_text in unique_texts:
            similarity = 100 * (1 - Levenshtein.distance(text, unique_text) / max(len(text), len(unique_text)))
            if similarity > threshold:
                indexes_to_remove.append(index)
                similarity_data.append({"text_1": unique_text, "text_2": text, "similarity": similarity})
                break
        else:
            unique_texts.append(text)

    df_filtered = df.drop(indexes_to_remove)
    df_similarities = pd.DataFrame(similarity_data)

    return df_filtered, df_similarities

test_cleaning = span_df.copy()

df_cleaned, df_similarities = remove_similar_text_fuzzy(test_cleaning, "text")
print(df_cleaned)
print(df_similarities)
```

```python
import spacy
import pandas as pd

# spaCy's English NLP model
nlp = spacy.load("en_core_web_sm")


df_keywords=text_df.copy()

def extract_keywords(text):
    doc = nlp(text)
    return [token.text for token in doc if token.is_alpha and not token.is_stop]

df_keywords['KEYWORDS'] = df_keywords['content'].apply(extract_keywords)

df_keywords[['heading','content', 'KEYWORDS']][:30]

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Combine all keywords into a single string per document
df_keywords['KEYWORDS_STR'] = df_keywords['KEYWORDS'].apply(lambda x: ' '.join(x))

df_keywords['KEYWORDS_STR'] = df_keywords['KEYWORDS_STR'].replace("", "empty")


#Compute TF-IDF weights
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df_keywords['KEYWORDS_STR'])

# Convert TF-IDF matrix to DataFrame for readability
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Add a column for weighted keywords (as a dictionary per document)
df_keywords['WEIGHTED_KEYWORDS'] = tfidf_df.apply(lambda row: {vectorizer.get_feature_names_out()[i]: weight

                                                       for i, weight in enumerate(row) if weight > 0}, axis=1)


pd.set_option("display.max_colwidth", None)

df_keywords[['heading','content', 'KEYWORDS','KEYWORDS_STR', 'WEIGHTED_KEYWORDS']][0:36]
```

```python
from collections import OrderedDict
from transformers import MPNetPreTrainedModel, MPNetModel, AutoTokenizer
import torch

# Mean Pooling - Take attention mask into account for correct averaging
def mean_pooling(model_output, attention_mask):
        token_embeddings = model_output #First element of model_output contains all token embeddings
        input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
        return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)

# Definition of ESGify class because of custom,sentence-transformers like, mean pooling function and classifier head
class ESGify(MPNetPreTrainedModel):
    """Model for Classification ESG risks from text."""

    def __init__(self,config): #tuning only the head
        """
        """
        super().__init__(config)
        # Instantiate Parts of model
        self.mpnet = MPNetModel(config,add_pooling_layer=False)
        self.id2label =  config.id2label
        self.label2id =  config.label2id
        self.classifier = torch.nn.Sequential(OrderedDict([('norm',torch.nn.BatchNorm1d(768)),
                                                ('linear',torch.nn.Linear(768,512)),
                                                ('act',torch.nn.ReLU()),
                                                ('batch_n',torch.nn.BatchNorm1d(512)),
                                                ('drop_class', torch.nn.Dropout(0.2)),
                                                ('class_l',torch.nn.Linear(512 ,47))]))


    def forward(self, input_ids, attention_mask):
         # Feed input to mpnet model
        outputs = self.mpnet(input_ids=input_ids,
                             attention_mask=attention_mask)

        # mean pooling dataset and eed input to classifier to compute logits
        logits = self.classifier( mean_pooling(outputs['last_hidden_state'],attention_mask))

        # apply sigmoid
        logits  = 1.0 / (1.0 + torch.exp(-logits))
        return logits
```

```python
import torch
import numpy as np
import pandas as pd
from transformers import AutoTokenizer

# Load model and tokenizer
model = ESGify.from_pretrained('ai-lab/ESGify')
tokenizer = AutoTokenizer.from_pretrained('ai-lab/ESGify')

# Ensure 'content' has no NaN values
df_classify = df_classify.dropna(subset=['content'])

# Tokenize the 'content' column
to_model = tokenizer.batch_encode_plus(
    df_classify['content'].tolist(),
    add_special_tokens=True,
    max_length=512,
    return_token_type_ids=False,
    padding="max_length",
    truncation=True,
    return_attention_mask=True,
    return_tensors='pt',
)

# Pass through the model
results = model(**to_model)

# Store predictions in separate lists
pred_labels = []
pred_scores = []

for idx in range(len(df_classify)):
    top_indices = torch.topk(results[idx], k=3).indices.tolist()
    top_classes = [model.id2label[i] for i in top_indices]
    top_scores = [np.round(results[idx, i].item(), 3) for i in top_indices]

    pred_labels.append(", ".join(top_classes))  # Store as a string
    pred_scores.append(top_scores)  # Store as a list

# Add structured predictions to the DataFrame
df_classify['pred_labels'] = pred_labels  # Text labels
df_classify['pred_scores'] = pred_scores  # Numerical scores

# Display updated DataFrame
print(df_classify)
```
