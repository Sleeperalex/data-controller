```python
import pandas as pd

# Charger le fichier avec diffrentes options
file_path = "controverses.csv"

try:
    # Essayer avec un sparateur par dfaut
    data = pd.read_csv(file_path, sep=',', error_bad_lines=False, warn_bad_lines=True)
except:
    # Essayer avec un point-virgule comme sparateur
    data = pd.read_csv(file_path, sep=';', error_bad_lines=False, warn_bad_lines=True)

# Afficher les premires lignes pour vrifier
print(data.head())
```

```python

# Afficher les donnes avant nettoyage
print("Avant nettoyage :")
print(data.info())
```

```python
# 1. Supprimer les doublons
data = data.drop_duplicates()
```

```python
# Taille avant suppression des doublons
initial_shape = data.shape[0]

# Supprimer les doublons
data = data.drop_duplicates()

# Taille aprs suppression des doublons
final_shape = data.shape[0]

# Vrification et affichage du rsultat
if initial_shape > final_shape:
    print(f"{initial_shape - final_shape} doublon(s) ont t supprim(s).")
else:
    print("Aucun doublon n'a t supprim.")
```

```python
def fill_missing_values(data):
    # Parcourir chaque colonne du DataFrame
    for col in data.columns:
        # Vrifier si la colonne est numrique
        if data[col].dtype in ['int64', 'float64']:
            # Remplir les valeurs manquantes avec 0 pour les colonnes numriques
            data[col] = data[col].fillna(0)
        # Vrifier si la colonne est de type texte (objet)
        elif data[col].dtype == 'object':
            # Remplir les valeurs manquantes avec 'Unknown' pour les colonnes de texte
            data[col] = data[col].fillna('Unknown')
    return data


# Remplir les valeurs manquantes
data = fill_missing_values(data)

# Afficher les premires lignes pour vrifier
print(data.head())

```

```python
# 3. Uniformiser les noms des colonnes (par exemple, en minuscules)
data.columns = [col.lower() for col in data.columns]
```

```python
# Afficher les noms de colonnes avant la modification
print("Avant uniformisation :")
print(data.columns)

# Uniformiser les noms des colonnes (les mettre en minuscules)
data.columns = [col.lower() for col in data.columns]

# Afficher les noms de colonnes aprs la modification
print("\nAprs uniformisation :")
print(data.columns)
```

```python
# 4. Nettoyer les colonnes de texte
# Supprimer les espaces autour des noms d'entreprises
data['company_name_mns'] = data['company_name_mns'].str.strip()
```

```python
# 5. Conversion des colonnes de type
# Conversion des dates (format YYYYMMDD)
data['date'] = pd.to_datetime(data['date'], format='%Y%m%d', errors='coerce')
```

```python
# Conversion des scores ESG en type float et validation des plages de valeurs
data['waste_env_score'] = pd.to_numeric(data['waste_env_score'], errors='coerce')
data['waste_env_score'] = data['waste_env_score'].clip(lower=0, upper=10)
```

```python
# Supprimer les lignes o des donnes critiques manquent (exemple : "company_name_mns")
data = data.dropna(subset=['company_name_mns'])
```

```python
# Afficher les donnes aprs nettoyage
print("\nAprs nettoyage :")
print(data.info())
print(data.head())
```

```python
# Sauvegarder les donnes nettoyes dans un nouveau fichier CSV
data.to_csv("controverses_cleaned.csv", index=False)
print("\nLes donnes nettoyes ont t sauvegardes dans 'controverses_cleaned.csv'.")
```
